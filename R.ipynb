```r
# Chargement des packages nécessaires
library(ggplot2)

# Création d'un jeu de données fictif pour les ventes mensuelles de quatre produits
data <- data.frame(
  Mois = rep(month.abb, times = 4),
  Ventes = c(150, 200, 250, 180, 210, 220, 230, 240, 260, 270, 280, 300, # Produit A
             100, 120, 140, 160, 180, 200, 220, 240, 260, 280, 300, 320, # Produit B
             80, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 190, # Produit C
             200, 220, 240, 260, 280, 300, 320, 340, 360, 380, 400, 420), # Produit D
  Produit = rep(c("Produit A", "Produit B", "Produit C", "Produit D"), each = 12)
)

# Création d'une palette de couleurs personnalisée
mes_couleurs <- c("Produit A" = "#FF5733", "Produit B" = "#FFC300", "Produit C" = "#DAF7A6", "Produit D" = "#C70039")

# Création du graphique à barres empilées avec ggplot2
ggplot(data, aes(x = Mois, y = Ventes, fill = Produit)) +
  geom_bar(stat = "identity", position = "stack") +
  scale_fill_manual(values = mes_couleurs) +
  labs(title = "Ventes Mensuelles par Produit",
       x = "Mois",
       y = "Ventes") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1))
```
